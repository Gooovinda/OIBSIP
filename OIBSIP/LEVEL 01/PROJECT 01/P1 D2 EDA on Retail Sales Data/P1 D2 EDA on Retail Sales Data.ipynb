# Exploratory Data Analysis on McDonald's Nutrition Data

This project analyzes the nutritional content of McDonald's menu items to uncover patterns and provide health-focused recommendations.


```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style='whitegrid')
```

```python
# loading the dataset
df = pd.read_csv("Nutrition Facts for McDonald's Menu.csv")
df.head()
```

* Data Cleaning
Check for missing values, duplicates, and data types.


```python
df.info()
df.isnull().sum()
df.duplicated().sum()
```

* Descriptive Statistics
 Get mean, median, mode, and standard deviation for key nutrients.


```python
df.describe()
```

* Category-wise Nutritional Analysis
Compare average nutrition per category.


```python
df.groupby('Category')[['Calories', 'Total Fat', 'Sugars', 'Protein', 'Sodium']].mean().sort_values(by='Calories', ascending=False)
```

* Top and Bottom Items
Identify highest and lowest calorie and sodium items.


```python
df.nlargest(5, 'Calories')[['Item', 'Category', 'Calories']]
df.nsmallest(5, 'Calories')[['Item', 'Category', 'Calories']]
df.nlargest(5, 'Sodium')[['Item', 'Category', 'Sodium']]
```

* Visualizations
Visualize average calories, distribution, and correlations.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Plot average calories by category
sns.barplot(data=df, x='Category', y='Calories', estimator=np.mean, errorbar=None)
plt.xticks(rotation=45)
plt.title('Average Calories by Category')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Calories'], bins=30, kde=True)
plt.title('Calories Distribution')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Between Nutritional Values')
plt.show()
```

* Final Insights
- Beverages have the highest average sugar content.
- Salads and Breakfast items are lower in calories compared to main meals.
- High correlation found between Calories and Total Fat.
- Sodium levels are extremely high in some items, especially sandwiches and combos.


** Actionable Recommendations 

* 1. Introduce Healthier Menu Options
   - Many items in the menu have high calories, especially in categories like Burgers and Sandwiches. Offering smaller portions or “light” versions could cater to calorie-conscious customers.

2. **Manage Sodium Levels**
   - Sodium levels in several items exceed daily recommended intake. McDonald's can reformulate recipes or offer low-sodium alternatives to appeal to health-aware consumers.

3. **Promote Balanced Meals**
   - Items with balanced nutrients (moderate fat, high protein, low sugar) should be highlighted as recommended picks on the menu.

4. **Provide Transparency on Sugar Content**
   - Beverages have very high sugar levels. Introduce low-sugar or sugar-free drink options, and include visual sugar warnings on the menu.

5. **Use Data-Driven Menu Labels**
   - Implement a “smart label” system showing calories, fat, sugar, and sodium per item to help customers make informed decisions.

6. **Target Marketing to Health-Conscious Demographics**
   - Promote Salads, Grilled Chicken, and Breakfast options as healthier choices in ads and mobile apps.
