```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading multiple CSV files
df1 = pd.read_csv('CAvideos.csv')
df2 = pd.read_csv('DEvideos.csv')
df3 = pd.read_csv('FRvideos.csv')
df4 = pd.read_csv('GBvideos.csv')
df5 = pd.read_csv('INvideos.csv')
df6 = pd.read_csv('JPvideos.csv', encoding='ISO-8859-1')
df7 = pd.read_csv('KRvideos.csv', encoding='ISO-8859-1')

df8 = pd.read_csv('MXvideos.csv', encoding='ISO-8859-1')

df9 = pd.read_csv('RUvideos.csv', encoding='ISO-8859-1')

df10 = pd.read_csv('USvideos.csv')

# Combine them into one DataFrame
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], ignore_index=True)

# Preview
df.head()
```

* Step 1: Load and Combine YouTube Trending Datasets from Multiple Countries
In this step:

We imported essential libraries: pandas, numpy, matplotlib, and seaborn.

We loaded 10 CSV files representing YouTube trending videos from different countries like CA, DE, FR, GB, IN, JP, KR, MX, RU, and US.

Some files were encoded in ISO-8859-1 to handle special characters (e.g., Japanese, Russian).

Finally, we combined all datasets into a single DataFrame using pd.concat() to prepare for unified analysis.

We also displayed the first few rows using .head() to verify successful loading.

```python
# Basic info
print(df.info())

# Check for missing values
print("\nMissing values:\n", df.isnull().sum())

# Check for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

* Step 2: Data Exploration â€“ Missing Values & Duplicates

The dataset has a total of 375,942 rows and 16 columns after combining all countries' video data.
We used .info() to check data types and found a mix of object, int64, and bool types.
Using .isnull().sum(), we discovered that only the description column has missing values (19,478 nulls), which is acceptable for text fields and can be optionally filled or dropped.
.duplicated().sum() showed 34,053 duplicate rows, which we can remove to ensure data integrity and avoid bias in our analysis.

```python
# Drop duplicate rows
df.drop_duplicates(inplace=True)

# Optionally fill missing descriptions with an empty string
df['description'].fillna('', inplace=True)

# Confirm cleanup
print("Duplicates after cleaning:", df.duplicated().sum())
print("Missing values after cleaning:\n", df.isnull().sum())
```

* Data Cleaning Summary
Duplicates Removed: Successfully removed 34,053 duplicate rows from the combined dataset.

Missing Values Handled: The only missing data was in the description column (19,478 entries), which has now been filled (likely using a placeholder or imputation).

Now, there are no missing values in any column.

The dataset is now clean and ready for analysis or visualization.

```python
df.to_csv('cleaned_youtube_data.csv', index=False)
```
